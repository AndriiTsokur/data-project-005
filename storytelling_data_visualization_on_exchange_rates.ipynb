# Storytelling Data Visualization on Exchange Rates

The dataset we'll use describes Euro daily exchange rates between April 1, 1999 (almost the very beginning of the introduction of this currency into circulation - January 1, 1999), and September 27, 2024 (the date of creation of this project). [The euro](https://en.wikipedia.org/wiki/Euro) (symbolized with €) is the official currency in most of the countries of the European Union.

[Daria Chemkaeva](https://www.kaggle.com/lsind18) put together the data set and made it available on [Kaggle](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020) — the data source is the European Central Bank. Note that the dataset gets regular updates.

The main goal of the project is to practice explanatory data visualization and storytelling, focusing on the exchange rate **between the euro and the American dollar**.

## Introducing the Dataset

```python
import numpy as np
import pandas as pd

exchange_rates = pd.read_csv('./euro-daily-hist_1999_2022.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

As we can see above, the dataset consists of 6,655 rows, each of which contains the euro exchange rate against 40 currencies (each column corresponds to a different currency), current as of the date specified in the first column. The dates specified in the dataset correspond exclusively to working days (Monday-Friday), on which banking institutions set the actual exchange rate of their national currency to euro.

About 25% of the currencies shown have more than 10% definitely missing data (`NaN`). In addition, 15% of currencies have 53% or more definitely missing data. Besides, there are cells with `-` symbol instead of `NaN`.

Most of the data types in the dataset are strings, so we will additionally need to convert most of them to data types appropriate for the project goals.

## Data Cleaning

Before we move on to the visualization process, we do the following:
* Rename the `[US dollar ]` and `Period\Unit:` columns to something easier to type - `us_dollar` and `date`.
* Change the `date` column to a `datetime` data type.
* Sort the values by `date` in ascending order.
* Reset the index, and drop the initial index.

```python
exchange_rates.rename(columns={'[US dollar ]': 'us_dollar', 'Period\\Unit:': 'date'}, inplace=True)
exchange_rates['date'] = pd.to_datetime(exchange_rates['date'])
exchange_rates.sort_values('date', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

Since we are only interested in the US dollar / euro exchange rate, let's separate the relevant columns into a new variable and examine it.

```python
euro_to_dollar = exchange_rates.copy()[['date', 'us_dollar']]
euro_to_dollar.info()
euro_to_dollar['us_dollar'].value_counts(dropna=False)
```

We found that in our `euro_to_dollar` dataset, sixty-two `-` characters are present instead of `NaN` values. At the same time, `NaN` are missing altogether. Thus, we have only 1% missing data - we can easily drop these strings. Then we need to convert the `us_dollar` column to the `float` data type.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['us_dollar'] != '-']
euro_to_dollar['us_dollar'] = euro_to_dollar['us_dollar'].astype(float)
euro_to_dollar.info()
```

## Rolling Mean

Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['date'], euro_to_dollar['us_dollar'])
plt.show()
```

We're going to use a 30-day rolling window to make the line smoother, increase the data-to-ink ratio, and focus the chart on long-term trends.

```python
euro_to_dollar['rolling_mean_30'] = euro_to_dollar['us_dollar'].rolling(30).mean()

plt.figure(figsize=(9, 3))

plt.subplot(1, 2, 1)
plt.plot(euro_to_dollar['date'], euro_to_dollar['us_dollar'])
plt.title('Original values', weight='bold')

plt.subplot(1, 2, 2)
plt.plot(euro_to_dollar['date'], euro_to_dollar['rolling_mean_30'])
plt.title('Rolling Window: 30', weight='bold')

plt.tight_layout()
plt.show()
```

## Storytelling Data Visualization

During the time period under consideration (1999-2024), there were many factors that influenced the change in the euro-dollar exchange rate.

To use the entire data range (1999-2024) and to practice in storytelling, we will take a possibility to visualize comparatively how the euro-dollar rate changed under the last four US presidents: George W. Bush (2001-2009), Barack Obama (2009-2017), Donald Trump (2017-2021) and Joe Biden (2021-2025).

```python
### Preparing a separate datasets for each president
presidents = euro_to_dollar.copy()[(euro_to_dollar['date'].dt.year >= 2001) & (euro_to_dollar['date'].dt.year < 2025)]
bush = presidents.copy()[presidents['date'].dt.year < 2009]
obama = presidents.copy()[(presidents['date'].dt.year >= 2009) & (
    presidents['date'].dt.year < 2017)]
trump = presidents.copy()[(presidents['date'].dt.year >= 2017) & (
    presidents['date'].dt.year < 2021)]
biden = presidents.copy()[presidents['date'].dt.year >= 2021]

### Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

### Adding the subplots
plt.figure(figsize=(16, 6))
ax1 = plt.subplot(2, 4, 1)
ax2 = plt.subplot(2, 4, 2)
ax3 = plt.subplot(2, 4, 3)
ax4 = plt.subplot(2, 4, 4)
ax5 = plt.subplot(2, 1, 2)
axes = [ax1, ax2, ax3, ax4, ax5]

### Changes to all the subplots
for ax in axes:
	ax.set_ylim(0.8, 1.7)
	ax.set_yticks([1.0, 1.2, 1.4, 1.6])
	ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6'], fontsize=11, alpha=0.3)
	ax.grid(alpha=0.5)

### ax1: Bush
ax1.plot(bush['date'], bush['rolling_mean_30'], color='#BF5FFF', linewidth=3)
ax1.text(12_390, 1.92, 'BUSH', fontsize=18, weight='bold', color='#BF5FFF')
ax1.text(12_100, 1.80, '(2001-2009)', weight='bold', alpha=0.3)
ax1.set_xticks(pd.date_range(start='2001', end='2009', freq='2YS'))
ax1.set_xticklabels(['2001', '2003', '2005', '2007', '2009'], fontsize=11, alpha=0.3)

### ax2: Obama
ax2.plot(obama['date'], obama['rolling_mean_30'], color='#FFA500', linewidth=3)
ax2.text(15_200, 1.92, 'OBAMA', fontsize=18, weight='bold', color='#FFA500')
ax2.text(15_020, 1.80, '(2009-2017)', weight='bold', alpha=0.3)
ax2.set_xticks(pd.date_range(start='2009', end='2017', freq='2YS'))
ax2.set_xticklabels(['2009', '2011', '2013', '2015', '2017'], fontsize=11, alpha=0.3)

### ax3: Trump
ax3.plot(trump['date'], trump['rolling_mean_30'], color='#00B2EE', linewidth=3)
ax3.text(17_650, 1.92, 'TRUMP', fontsize=18, weight='bold', color='#00B2EE')
ax3.text(17_560, 1.80, '(2017-2021)', weight='bold', alpha=0.3)
ax3.set_xticks(pd.date_range(start='2017', end='2021', freq='1YS'))
ax3.set_xticklabels(['2017', '2018', '2019', '2020', '2021'], fontsize=11, alpha=0.3)

### ax4: Biden
ax4.plot(biden['date'], biden['rolling_mean_30'], color='#28A745', linewidth=3)
ax4.text(19_170, 1.92, 'BIDEN', fontsize=18, weight='bold', color='#28A745')
ax4.text(19_040, 1.80, '(2021-2025)', weight='bold', alpha=0.3)
ax4.set_xticks(pd.date_range(start='2021', end='2025', freq='1YS'))
ax4.set_xticklabels(['2021', '2022', '2023', '2024', '2025'], fontsize=11, alpha=0.3)

### ax5: Bush-Obama-Trump-Biden
ax5.plot(bush['date'], bush['rolling_mean_30'], color='#BF5FFF', linewidth=3)
ax5.plot(obama['date'], obama['rolling_mean_30'], color='#FFA500', linewidth=3)
ax5.plot(trump['date'], trump['rolling_mean_30'], color='#00B2EE', linewidth=3)
ax5.plot(biden['date'], biden['rolling_mean_30'], color='#28A745', linewidth=3)
ax5.set_xticks([])

### Adding a title and a subtitle
ax1.text(10_900, 2.35, 'EUR / USD rate averaged 1.19 under the last four US presidents', fontsize=24, weight='bold')
ax1.text(10_900, 2.11, '''EUR / USD exchange rates under George W. Bush (2001-2009), Barack Obama (2009-2017), 
Donald Trump (2017-2021) and Joe Biden (2021-2025)''', fontsize=16, alpha=0.7)

### Adding a signature
ax5.text(10_700, 0.65, '©DATAQUEST' + ' '*160 + 'Source: European Central Bank', color='#F0F0F0', backgroundcolor='#4D4D4D', size=14)

plt.show()
```

## Conclusion

Analysis of the data shows that a combination of external and internal factors, as well as the policies chosen by the US during the presidencies of Barack Obama and Joe Biden, contributed most to the strengthening of the US dollar against the euro. 

Identifying the specific reasons that contributed to this requires more detailed analysis, taking into account the events that, in one way or another, may have influenced this process.
